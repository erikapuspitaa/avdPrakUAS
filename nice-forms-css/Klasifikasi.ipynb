```python
df = data.rename(columns={
    '1. Apakah Anda bermain game online?': 'bermain_game',
    '2. Seberapa sering Anda bermain game online dalam sebulan terakhir?': 'durasi_bermain',
    '3. Berapa lama waktu yang Anda habiskan setiap hari untuk bermain game online? ' : 'durasi_waktu_bermain',
    '4. Apa jenis game online yang paling sering Anda mainkan? (Contoh: MOBA, FPS, RPG, dll.)' : 'jenis_game',
    '5. Apakah Anda merasa bermain game online mengganggu waktu belajar atau aktivitas Anda?' : 'game_mengganggu_aktivitas',
    '6. Bagaimana Anda menilai tingkat rasa stres Anda sebulan terakhir? ' : 'tingkat_stress',
    '7.  Seberapa sering Anda merasa cemas, depresi, tertekan, atau memiliki masalah kesehatan mental lainnya dalam sebulan terakhir?' : 'frekuensi_kesehatan_mental',
    '8. Apa yang menjadi penyebab utama stres Anda sebulan terakhir?' : 'penyebab_utama_stress',
    '9. Apakah bermain game online menjadi pilihan pertama Anda ketika Anda merasa stres? atau apakah Anda memiliki metode lain yang Anda coba terlebih dahulu?' : 'pilihan_pertama_stress',
    '10. Jika YA, berikan alasan mu. Jika TIDAK, aktivitas apa yang menjadi pilihan pertamamu ketika merasa stres?' : 'alasan_pilihan_aktivitas',
    '11. Apakah Anda merasa bahwa bermain game online telah membantu Anda mengalihkan perhatian dari sumber stres yang sedang Anda alami?' : 'membantu_mengalihkan_stress',
    '12. Apakah Anda pernah mengalami situasi di mana bermain game untuk menghilangkan stres malah menambah stres atau masalah lain dalam hidup Anda?' : 'game_menambah_stress',
    '13. Bagaimana pendapat Anda tentang sejauh mana bermain game sebagai cara untuk menghilangkan stres merupakan alternatif yang sehat dan efektif? ' : 'klasifikasi'})
```

```python
df['Jenis Kelamin']=np.where(df['Jenis Kelamin']=='Laki-laki',0,1)
df['bermain_game']=np.where(df['bermain_game']=='Tidak',0,1)
df['Pendidikan'] = np.where(df['Pendidikan'] == 'Pelajar', 0,
                                           np.where(df['Pendidikan'] == 'Mahasiswa', 1,
                                                    np.where(df['Pendidikan'] == 'Umum / Sudah bekerja', 2, -1)))
df['durasi_bermain'] = np.where(df['durasi_bermain'] == 'Setiap hari', 4,
                                           np.where(df['durasi_bermain'] == 'Beberapa kali seminggu', 3,
                                                    np.where(df['durasi_bermain'] == 'Sekali seminggu', 2,
                                                            np.where(df['durasi_bermain'] == 'Jarang', 1,
                                                                     np.where(df['durasi_bermain'] == 'Tidak pernah', 0,-1)))))
df['durasi_waktu_bermain'] = np.where(df['durasi_waktu_bermain'] == 'Lebih dari 2 jam', 3,
                                           np.where(df['durasi_waktu_bermain'] == '1-2 jam', 2,
                                                    np.where(df['durasi_waktu_bermain'] == '30 menit - 1 jam', 1,
                                                            np.where(df['durasi_waktu_bermain'] == 'Kurang dari 30 menit', 0,-1))))
df['game_mengganggu_aktivitas'] = np.where(df['game_mengganggu_aktivitas'] == 'Tidak', 0,
                                           np.where(df['game_mengganggu_aktivitas'] == 'Ya', 2,
                                                    np.where(df['game_mengganggu_aktivitas'] == 'Netral', 1, -1)))
df['tingkat_stress'] = np.where(df['tingkat_stress'] == 'Sangat Tinggi', 4,
                                           np.where(df['tingkat_stress'] == 'Tinggi', 3,
                                                    np.where(df['tingkat_stress'] == 'Normal', 2,
                                                            np.where(df['tingkat_stress'] == 'Rendah', 1,
                                                                     np.where(df['tingkat_stress'] == 'Sangat Rendah', 0,-1)))))
df['frekuensi_kesehatan_mental'] = np.where(df['frekuensi_kesehatan_mental'] == 'Tidak Pernah', 0,
                                           np.where(df['frekuensi_kesehatan_mental'] == 'Jarang', 1,
                                                    np.where(df['frekuensi_kesehatan_mental'] == 'Normal', 2,
                                                             np.where(df['frekuensi_kesehatan_mental'] == 'Sering', 3,
                                                                      np.where(df['frekuensi_kesehatan_mental'] == 'Sangat Sering', 4,-1)))))
df['pilihan_pertama_stress']=np.where(df['pilihan_pertama_stress']=='Tidak',0,1)
df['membantu_mengalihkan_stress'] = np.where(df['membantu_mengalihkan_stress'] == 'Sangat membantu', 4,
                                           np.where(df['membantu_mengalihkan_stress'] == 'Membantu', 3,
                                                    np.where(df['membantu_mengalihkan_stress'] == 'Netral', 2,
                                                            np.where(df['membantu_mengalihkan_stress'] == 'Kurang membantu', 1,
                                                                     np.where(df['membantu_mengalihkan_stress'] == 'Tidak membantu', 0,-1)))))
df['game_menambah_stress'] = np.where(df['game_menambah_stress'] == 'Tidak pernah', 0,
                                           np.where(df['game_menambah_stress'] == 'Jarang', 1,
                                                    np.where(df['game_menambah_stress'] == 'Normal', 2,
                                                             np.where(df['game_menambah_stress'] == 'Sering', 3,
                                                                      np.where(df['game_menambah_stress'] == 'Sangat sering', 4,-1)))))
df['klasifikasi'] = np.where(df['klasifikasi'] == 'Sangat setuju', 3,
                              np.where(df['klasifikasi'] == 'Setuju', 2,
                                          np.where(df['klasifikasi'] == 'Netral', 1,
                                                      np.where(df['klasifikasi'] == 'Tidak Setuju', 0, -1))))
```

# **Train Data**

```python
df=df.drop('jenis_game',axis=1)
```

```python
df=df.drop('penyebab_utama_stress',axis=1)
```

```python
df=df.drop('alasan_pilihan_aktivitas',axis=1)
```

```python
num = ['membantu_mengalihkan_stress','pilihan_pertama_stress','durasi_bermain','durasi_waktu_bermain','bermain_game']
```

```python
df[num]
```

```python
y=df['klasifikasi']
x=df[num]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
# divided into train data and test data with proportion in 70% (train) and 30% (test)
# the form will be: x_train, y_train, x_test, and y_test
```

```python
x_train
```

# **Modelling**

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Random Forest
clf = RandomForestClassifier(max_depth=5, n_estimators=100, random_state=42)  # Ganti kernel dan parameter sesuai kebutuhan

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

# #Predict the response for test dataset
rf_pred = clf.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print(f'Akurasi Random Forest: {rf_accuracy * 100:.2f}%')
```

```python
print(classification_report(rf_pred,y_test))
```

```python
import pickle
# Simpan model ke file
with open('model.pkl', 'wb') as model_file:
    pickle.dump(clf, model_file)
```
